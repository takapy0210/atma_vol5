## クラスタリング

```python
from sklearn.cluster import KMeans
from sklearn import metrics
import statistics
```

```python
scores = []
values = np.arange(2, 50)
def kmeans(n_clust, df):
    km = KMeans(n_clusters=n_clust, init="k-means++", random_state=0)
    y_km = km.fit_predict(df)
    return y_km

def n_label(y_km):
    cluster_labels = np.unique(y_km)
    n_clusters = cluster_labels.shape[0]
    return [cluster_labels,n_clusters]
```

## UMAP

```python
%%time
# クラスタ数を変化させながら、k-meansを生成して訓練を繰り返す（PCA前のデータに対して実施）
for num_clusters in values:
    y_km = kmeans(num_clusters, umap_df)
    nn_l = n_label(y_km)
    cluster_labels = nn_l[0]
    n_clusters = nn_l[1]
    
    #シルエット係数を計算
    silhouette_vals = metrics.silhouette_score(umap_df, y_km, metric='euclidean', sample_size=int(len(umap_df)/2))
    
    print('silhouette_score =', silhouette_vals)
    
    sil = []  #クラスタ内のサンプル数を格納する配列
    for i, c in enumerate(cluster_labels):
        c_silhouette_vals = y_km.tolist().count(c)
        sil.append(c_silhouette_vals)
    
    print('sample:', sil)
    print('k:', int(i+1))
    print('std:', statistics.pstdev(sil))
    print('----------')
```
