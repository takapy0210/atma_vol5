```python
import pandas as pd
import os
```

```python
DIR = './score'
sorted(os.listdir(DIR))
```

```python
def score_mean(df):
    df = df.groupby('run_name').mean().round(4).reset_index().sort_values('run_name')
    return df
```

```python
score1 = pd.read_csv(DIR + '/' + 'LB08771_lgb_StratifiedKFold_06051336_score.csv')
score2 = pd.read_csv(DIR + '/' + 'cat_StratifiedKFold_06061151depth5_pl_score.csv')
score3 = pd.read_csv(DIR + '/' + 'lgb_StratifiedKFold_06061146_score.csv')
score4 = pd.read_csv(DIR + '/' + 'lgb_StratifiedKFold_06061325cat_lgbm_pl45_55_add_score.csv')
```

```python
score1 = score_mean(score1)
score2 = score_mean(score2)
score3 = score_mean(score3)
score4 = score_mean(score4)
```

```python
score = pd.merge(score1, score2, on='run_name', how='left')
score = pd.merge(score, score3, on='run_name', how='left')
score = pd.merge(score, score4, on='run_name', how='left')
```

```python
score = score.T
score.columns = score.iloc[0]
score = score.drop(score.index[0])
```
